```python
import os
import json

from tools.data_loader import import_database
from tools.accessions_db import AccessionDatabase
from tools.accession import Accession

RAW_DATA = 'data/raw'
DEFAULT_DIRS = ('Definite MG','Healthy control')


imported_dataset = import_database(RAW_DATA, DEFAULT_DIRS)
AccDatabase = AccessionDatabase(imported_dataset, DEFAULT_DIRS)
```

```python
# Analyze the accessions
accession_samples = {k:AccDatabase.get_accession(k) for k in AccDatabase.accessions.keys()}
for p_idx, (k, acc) in enumerate(accession_samples.items()):
    acc : Accession = acc.analyse(Accession.DEFAULT_ANALYSIS) # gets gain, velocity, and speed
```

```python
patient_holder = {}
for acc_idx, acc_data in enumerate(accession_samples.values()):
    # if len(normalised_patient_holder)>1:
    #     break
    acc_data: Accession
    acc_info =  acc_data.accession

    print(f'({acc_idx+1}/{len(accession_samples)})  STARTING {acc_info}')
    axis_info = acc_data.axis
    patient_id = '-'.join(f'{d:02d}' for d in acc_info[:2])
    freq_info = 'freq_' + str([0.5,0.75,1.0][acc_info[-1]])
    point_amt = [480,360,240][acc_info[-1]]

    # top layer
    default_data = {
        'patient_name':acc_data.patient_name,
        'pathology_class':acc_data.pathology_class,
        'stored':[],
        'visits':{}
    }
    patient_holder[patient_id] = patient_holder.setdefault(patient_id, default_data)

    if acc_info in patient_holder[patient_id]['stored']:
        print('└─ SKIPPED, ALREADY IN STORAGE','+'*30,sep="\n")
        continue

    top_layer = patient_holder[patient_id]

    visit_data = {
        'visit_date':acc_data.visit_date,
    }

    middle_layer:dict

    # find visit for middle layer
    for visit_idx, visit in top_layer['visits'].items():
        if visit['visit_date'] == acc_data.visit_date:
            middle_layer = top_layer['visits'][visit_idx]
            break

    else:
        visit_idx = len(top_layer['visits'])
        top_layer['visits'][visit_idx] = visit_data
        middle_layer = top_layer['visits'][visit_idx]

    middle_layer[freq_info] = middle_layer.setdefault(freq_info,{})
    final_layer = middle_layer[freq_info]

    # store data in final layer
    wanted_data = {
        f'velocity_{axis_info}':acc_data.velocity_dict,
        f'speed_{axis_info}':acc_data.speed_dict,
        f'gain_{axis_info}':acc_data.gain_dict,
    }
    
    for k,v in wanted_data.items():
        for attr_k, attr_v in v.items():
            final_layer[k+attr_k] = attr_v

    top_layer['stored'].append(acc_info)
    
    print(f'└─ FINISHED {acc_info}','+'*30,sep="\n")
```

```python
patient_count = len(patient_holder)
file_count = sum([len(p['stored']) for p in patient_holder.values()])
print(f'Patient count: {patient_count}, file_count: {file_count}')

try:
    os.mkdir('./out')
    os.mkdir('./out/dataset')
except:
    pass

export_path = './out/dataset'
```

```python
for p_idx, (patient_k, patient_v) in enumerate(patient_holder.items()):
    use_path = os.path.join(export_path, patient_v['pathology_class'])
    if not os.path.isdir(use_path):
        os.mkdir(use_path)
    file_name = os.path.join(use_path,f'{patient_k}.json')
    with open(file_name,'w',encoding='utf-8') as out_file:
        json.dump(patient_v,out_file,indent=2)
    print(f'{p_idx+1}/{len(patient_holder)} at {file_name}')
```
